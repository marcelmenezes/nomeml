```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.metrics import make_scorer
from sklearn.tree import DecisionTreeRegressor 
from sklearn.model_selection import ShuffleSplit, GridSearchCV


# O fit_model(...) cria uma um estimador de Ã¡rvore de decisÃ£o, faz o CV do parÃ¢metro 'max_depth' de 1 a 10, depois faz um Grid Search CV, e retorna o grid.best_estimator_;
def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    cv_sets = ShuffleSplit(n_splits=5, test_size = 0.20, random_state = 0)

    # TODO: Create a decision tree regressor object
    regressor = RandomForestClassifier()

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    #params = {"n_estimators":[10, 20, 40, 80, 200]}

    param_grid = { 
        #'n_estimators': [200, 500, 2000],
        'n_estimators': [200],
        #'max_features': ['auto', 'sqrt', 'log2'],
        'max_depth' : [12,20],
        'max_depth' : [12],
        'criterion' :['gini', 'entropy']
        #'criterion' :['gini']
    }
    
    # TODO: Create the grid search object
    grid = GridSearchCV(regressor, param_grid, cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

```python
param_grid = {
    'n_estimators': 200,
    'max_depth' : 12,
    'criterion' :'gini'
}


# Fit the training data to the model using grid search
#est = fit_model(X_train, y_train)
est = RandomForestClassifier(**param_grid)
est.fit(X, y)

print(est.get_params())
```
